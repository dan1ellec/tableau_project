```python
import numpy as np
import pandas as pd
import os
```

```python
#loading csvs into a df dictionary
folder = 'JC_data/'
JC_dataframes = {}

for file in os.listdir(folder):
    filename = os.fsdecode(file)
    if filename.endswith('.csv'):
        JC_dataframes[filename.split('.')[0]] = pd.read_csv(os.path.join(folder, filename)) 
```
